## 挑战A-1：如法泡制

* 先国再城

````python
# A-1 原完整代码 "上次的EXCEL表的python产出代码"
先国再行 = df.groupby ( by = ['国家','行业'] ) \
             .agg ({ "企业名称" : "count", \
                     "估值（亿人民币）":["sum","mean"], \
                     "成立年份":["max","min"],               }) \
             .sort_values ( by = [("估值（亿人民币）","sum")], ascending = False) \
             .rename ( columns = {"sum":"总和", "mean":"均值", "count":"数量", "max":"最新", "min":"最早"} )
先行再国 = df.groupby ( by = ['行业', '国家'] ) \
             .agg ({ "企业名称" : "count", \
                     "估值（亿人民币）":["sum","mean"], \
                     "成立年份":["max","min"],               }) \
             .sort_values ( by = [("估值（亿人民币）","sum")], ascending = False) \
             .rename ( columns = {"sum":"总和", "mean":"均值", "count":"数量", "max":"最新", "min":"最早"} )
display(先国再行)
display(先行再国)

with pd.ExcelWriter("20春_pandas_week03_hurun_unicorn.xlsx") as writer:
    先国再行.to_excel(writer,sheet_name="先国再行") 
    先行再国.to_excel(writer,sheet_name="先行再国") 
````

## 聚合B-1：单分组单列（推荐写法）

agg方法参数，我们强力推荐用数据科学家眼光，从表格的

1.**变量** 为 键
2.**计算量** 为 值，可一个可多个序列

来做成字典

````python
# B-1 推荐作法：聚合操作用的参数 从 A-1 代码取出如下
agg方法参数 =      { "企业名称" : "count", \
                     "估值（亿人民币）":["sum","mean"], \
                     "成立年份":["max","min"],           }

# 学习的方式，先从简单开始，化繁为简
# 怎麽只做变量"企业名称"的（单组单列聚合）
聚合_字典_簡單 =  df.groupby("行业")\
                   .agg({ "企业名称" : "count" })
#                           ^^^变量     ^^^计算量
display ( 聚合_字典_簡單.head() )
````

## 聚合B-1 Extra：不推荐但需读懂的写法

agg方法参数，也有其它写法，可不用字典用列表，但表达力比较不足，不推荐使用。

但仍需读懂并知其特性：

注意到了吗？要对要处理的变量**先进行切片**，如 以下代码的[["企业名称"]]

这种作法比较不能用在**复杂的聚合**，只能挑取部分变量，然后要进行一致的聚合计算量，比较没有弹性，虽然代码看来比较简单，但不是数据分析/科学思维。

为何不直接整合呢？所以我们不推荐先切片再列表聚合的方式

````python
# B-1 Extra：不推荐但需读懂的写法

聚合_列表_agg = df.groupby("行业")\
                  [["企业名称"]]\
                  .agg(["count"])
display ( 聚合_列表_agg.head() )

# 连agg但不用，直接用count()
聚合_列表_count = df.groupby("行业")\
                   [["企业名称"]]\
                   .count()
display ( 聚合_列表_count.head() )
````

## 聚合B-2：单分组多列 (推荐写法)

agg方法参数，当选取多列进行不同的运算时：

1.**变量** 为 键  
2.**计算量** 为 值

不同的 **键** (列)  

我们可以用不同的 **计算量** (值)存为字典

````python
# B-2 聚合操作：单分组多列 每列有自己的agg方法
df.groupby("国家")\
    .agg ({ "企业名称" : "count", \
            "估值（亿人民币）":["sum","mean"], \
            "成立年份":["max","min"],               
          }) 
````

## 聚合B-2 Extra1：不推荐但需读懂的其它写法
* 这里大家在来看，如果先切片：，进行多列的运算，可能很难实现每列按照不同的运算进行，这就更加体现了数据科学的B-1推荐方法

````python
# B-2 Extra 聚合操作：单组多列 多列同样的agg

# 计算多个不同列的多个 函数值 df_grouped_03:(["国家","行业"])
df.groupby("国家")[["估值（亿人民币）","成立年份"]].agg(["max","min"])
````

## 聚合B-2 Extra2： agg 的几种方式（多指标统计的方法）

我们需要掌握推荐的agg作法，也需要看懂不同的使用方法，这便于我们更好的阅读 **不同人** 写的代码

```python   

df.groupby(["国家","行业"])["估值（亿人民币）"].agg(sum = "sum",mean = "mean",count = "count")
df.groupby(["国家","行业"])["估值（亿人民币）"].agg(["sum","mean","max","min"])
df.groupby(["国家","行业"]).agg({"估值（亿人民币）":["sum","mean","count"]})
```
