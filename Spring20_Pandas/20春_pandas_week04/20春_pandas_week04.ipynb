# 数据框重塑
##  Pandas的stack实现数据透视

stack()是将原来的列索引转成了最内层的行索引，这里是多层次索引，其中AB索引对应第三层，即最内层索引。
![stack](20春_pandas_week04_20春_pandas_week04_stack_01.png)

* 首先我们可以从表中发现 **"掌门人/创始人"** 和 **"部分投资机构"** 可能不为1
* 这意味涉及到每一个独角兽企业会有**多个创始人**、**多个投资商**
* 为下一节merge合并表格作准备
    * 1、创建表："企业名称"——>"掌门人/创始人" 构建1对多列表
    * 2、创建表："企业名称"——>"部分投资机构" 构建1对多列表
    
* 附加：思考，如果想用投资方作为一对多的一，应该如何实现？
    * 3、创建表："投资方"——>"企业名称"
    
-----

* 本节具体操作：
    * 1、完整代码直接来看效果～
    * 2、分布讲解：drop(易)、split(易)、stack(难点)、merge(初次尝试，下节具体讲解)
    

```python
# B-1 本节代码：
df_独角兽_创始人_准备 =df[["企业名称","掌门人/创始人"]]
df_独角兽_创始人_准备.index.name="序号"
df_独角兽_创始人 =pd.merge(df_独角兽_创始人_准备.drop('掌门人/创始人', axis=1),\
                          df['掌门人/创始人'].str.split(',', expand=True)\
                                            .stack().reset_index(level=1,drop=True)
                                            .rename('掌门人/创始人'),\
                          on="序号")

df_独角兽_投资机构_准备 = df[["企业名称","部分投资机构"]]
df_独角兽_投资机构_准备.index.name="序号"
df_独角兽_投资机构 = pd.merge(df_独角兽_投资机构_准备.drop('部分投资机构', axis=1),\
                             df['部分投资机构'].str.split('、', expand=True)\
                                             .stack().reset_index(level=1, drop=True)
                                             .rename('部分投资机构'),\
                             on="序号")


display(df_独角兽_创始人)
display(df_独角兽_投资机构)
```

### 练习：你能创建表："投资方"——>"企业名称"进行一对多的练习么？

* 本节提供完整代码
* 本节重难点：contains、列表推到式

```python
# C-1 直接给完整代码：
df_投资方_企业名称 = df[["部分投资机构","企业名称"]]
所有投资机构 = [item for sublist in df_投资方_企业名称["部分投资机构"].str.split('[、,,]') for item in sublist]

touzi_dict = {}
for item in 所有投资机构:
    df_投资方_item = df_投资方_企业名称[df_投资方_企业名称["部分投资机构"].str.contains(item)]
    touzi_dict[item] =','.join([i for i in df_投资方_item["企业名称"].values])

投资方_独角兽企业_Series = pd.Series(touzi_dict)

df_投资方_独角兽企业 = pd.DataFrame({"投资方":投资方_独角兽企业_Series.index,"独角兽企业":投资方_独角兽企业_Series.values})
df_投资方_独角兽企业.index.name="序号"
df_投资方_独角兽企业 =pd.merge(df_投资方_独角兽企业.drop('独角兽企业', axis=1),\
                     df_投资方_独角兽企业['独角兽企业'].str.split(',', expand=True).stack().reset_index(level=1,drop=True).rename('独角兽企业'),\
                     on="序号")
df_投资方_独角兽企业
```

#### 分布详细讲解

* C-2 准备投资方->企业名称表格
* C-3 尝试找到“某一个”投资方所投资的企业有哪些，并以投资方为key，投资企业名称为values构建字典
* C-4 构建所有的投资方对应投资企业的字典

```python
# C-3 尝试找到“某一个”投资方所投资的企业有哪些，并以投资方为key，投资企业名称为values构建字典
touzi_dict_test = {}
df_投资方_春华资本 = df_投资方_企业名称[df_投资方_企业名称["部分投资机构"].str.contains("春华资本")]
touzi_dict_test["春华资本"] =df_投资方_春华资本["企业名称"].values
display(df_投资方_春华资本)
touzi_dict_test
```

```python
# C-4 构建所有的投资方对应投资企业的字典
所有投资机构 = [item for sublist in df_投资方_企业名称["部分投资机构"].str.split('[、,,]') for item in sublist]
所有投资机构
touzi_dict = {}
for item in 所有投资机构:
    df_投资方_item = df_投资方_企业名称[df_投资方_企业名称["部分投资机构"].str.contains(item)]
    touzi_dict[item] =','.join([i for i in df_投资方_item["企业名称"].values])
touzi_dict
```

```python
# C-5 转化为Dataframe
投资方_独角兽企业_Series = pd.Series(touzi_dict)
display(投资方_独角兽企业_Series)
df_投资方_独角兽企业 = pd.DataFrame({"投资方":投资方_独角兽企业_Series.index,"独角兽企业":投资方_独角兽企业_Series.values})
display(df_投资方_独角兽企业)
```
